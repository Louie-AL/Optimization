**Example Code**

You may find useful the following code.  It solves a simple linear program, and finds sensitivity information.

```python
# Define Variables
# Rather than requiring "nonneg=True" for the variables,
# we instead use explicit constraints (a few lines down in the code),
# in order to access sensitivity information.
x1 = cp.Variable(name='x1')
x2 = cp.Variable(name='x2')

# Objective
objective = x1 + x2

# Constraints
constraints = []
constraints += [x1 <= 2]
constraints += [x1 + 2*x2 <= 4]
constraints += [x1 >= 0] # Explicitly enforce non-negativity
constraints += [x2 >= 0]

# Define problem
problem = cp.Problem(cp.Maximize(objective), constraints)

# Solve problem
problem.solve(solver=cp.GUROBI, verbose=False)
print('objective %s' % problem.value)

# Print variable values
for variable in problem.variables():
    print("%s = %s" % (variable.name(), variable.value))

# Print sensitivity
print('Sensitivity \t Associated-Constraint')
for cons in constraints:
  print("%s \t\t %s"%(cons.dual_value, cons))
```
